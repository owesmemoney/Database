### Task 2: Complicated Queries

__Note:__ Performing the query in this task may require changing column values or types.

#### Query - Career Summary

- Write a query that produces a result of the form:
    - ```playerID```
    - ```nameLast```
    - ```nameFirst```
    - The sum of ```appearances.G_all``` for the player over all rows.
    - The sum over all rows of the following columns from batting:
        - ```G```
        - ```AB```
        - ```R```
        - ```AB```
        - ```2B```
        - ```3B```
        - ```HR```
        - ```RBI```
        - ```BB```
  
    - ```batting_average```, which is defined as $
\dfrac{sum(H)}{sum(AB)}
$

    - ```on_base_percentage```, which is defined as $
\dfrac{(sum(H) + sum(BB))}{(sum(AB) + sum(BB))}
$


- The query should be limited to 20 rows, and sorted by ```on_base_percentage``` from highest to lowest.


- ```batting_average``` and ```on_base_percentage``` should round to three decimal places.

````python
%%sql with people_basic as
    (
        select playerID, nameLast, nameFirst from people
    ),
appearances_career as
         (
             select playerID, sum(G_all) as total_g from appearances group by playerID limit 20
         ),
basic_appearance as (
    select *from people_basic join appearances_career using(playerID)
),

batting_G as (
    select playerID, sum(G) as sum_batting_G from batting group by playerID limit 20
),
rows_sumg as (
select *from basic_appearance join batting_G using(playerID)
    ),

batting_AB as (
    select playerID, sum(AB) as sum_AB from batting group by playerID limit 20
),
rows_sumAB as (
select *from rows_sumg join batting_AB using(playerID)
    ),

batting_R as (
    select playerID, sum(R) as sum_R from batting group by playerID limit 20
),
rows_sumR as (
select *from rows_sumAB join batting_R using(playerID)
    ),

batting_2B as (
    select playerID, sum(2B) as sum_2B from batting group by playerID limit 20
),
rows_sum2B as (
select *from rows_sumR join batting_2b using(playerID)
    ),
batting_3B as (
    select playerID, sum(3B) as sum_3B from batting group by playerID limit 20
),
rows_sum3B as (
select *from rows_sum2B join batting_3B using(playerID)
    ),

batting_HR as (
    select playerID, sum(HR) as sum_HR from batting group by playerID limit 20
),
rows_sumHR as (
select *from rows_sum3B join batting_HR using(playerID)
    ),

batting_RBI as (
    select playerID, sum(RBI) as sum_RBI from batting group by playerID limit 20
),
rows_sumRBI as (
select *from rows_sumHR join batting_RBI using(playerID)
    ),

batting_BB as (
    select playerID, sum(BB) as sum_BB from batting group by playerID limit 20
),
rows_sumBB as (
select *from rows_sumRBI join batting_BB using(playerID)
    ),

batting_average as (
    select playerID, round(sum(H)/if(sum(AB)=0,NULL, sum(AB)),3) as batting_average from batting group by playerID limit 20

    ),
average_sums as (
select * from rows_sumBB join batting_average using(playerID)
    ),
on_base_percentage as (
select playerID, round((sum(H)+sum(BB))/if ((sum(AB)+sum(BB))=0,NULL, ((sum(AB)+sum(BB)))),3) as on_base_percentage from batting group by playerID limit 20
    )
select *from average_sums join on_base_percentage using(playerID);


````
